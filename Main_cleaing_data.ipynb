```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Dima\Downloads\Jupyter_projectes\Untitled Folder 3\Cleaing_dataframe_and_data_handling\final_df.csv",dtype={'col':np.float})
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.rename(columns={'Unnamed: 0.1':'Causes'},inplace=True)
df.set_index('Causes',inplace=True)
df=df.T
```

```python
for i in df.columns:
    df.set_index(i)

df.dropna(axis=0, thresh=150,inplace=True)
df.dropna(axis=1, thresh=163,inplace=True)
list_a=df.columns.tolist()
```

```python

list_removed=['All Cancers','CAPE Ratio by Country','Other Injuries',
             'Current Account Balance ','Mobile Cellular Telephones ','Land Area ', 'Coastline Lengths ',
             'Highest Points ','Exports ','Death Rates ',
             'Crude Oil Production ','Electrification ','Urban Population ','Stabbing Deaths by Country',
             'Cheapest Countries To Live in','Fastest Growing Economies',
             'List of Tariffs by Country','Major Countries Currencies','Most Expensive Countries To Live in',
             '911 by Country','Death Rate by Country'
             ,  'Net Migration Rates ','Natural Gas Exports ','Urban Population Growth Rates ','Coronavirus by Country'
             ,'Airports ','Other Neoplasms','Internet Users ','Infant Mortality Rate by Country','Birth Life Expectancy ']
```

```python
df.drop(columns=list_a[7],axis=1,inplace=True)
for i in list_removed:
    df.drop(columns=i,axis=1,inplace=True)
```

```python

df = df.replace(',','', regex=True)
df = df.replace('-',np.nan, regex=True)
df = df.replace('--',np.nan, regex=True)
df
```

```python
df= df.astype(str)



for i in df.columns:
    if(  (df[i].iat[0]).isnumeric()==True ):
        df[i]=df[i].astype(float)
        
    else:
            if('%' in (df[i].iat[0]) ):
                df[i]=df[i].str.strip("%")
                df[i]=df[i].astype(float)/100
            elif('$' in (df[i].iat[0]) ):
                for s in range(0,len(df[i].values)):
                    df[i].values[s]=df[i].values[s].strip("$")
                    if ('Bn' in df[i].values[s] ):
                        df[i].values[s]=df[i].values[s].strip("Bn")
                        df[i].values[s]=float(df[i].values[s])*pow(10,9)
                    elif('Tn' in df[i].values[s]  ):
                        df[i].values[s]=df[i].values[s].strip("Tn")
                        df[i].values[s]=float(df[i].values[s])*pow(10,12)
                    elif('Mn' in df[i].values[s]  ):
                        df[i].values[s]=df[i].values[s].strip("Mn")
                        df[i].values[s]=float(df[i].values[s])*pow(10,6)
                df[i]=df[i].astype(float)
              
                    
df = df.replace('NaN',np.nan, regex=True)
df = df.replace('nan',np.nan, regex=True)
```

```python
dict_nulls={}
index_numbers={}

for i in df.columns:
    if(df[i].isna().sum()>0):
        dict_nulls[df[i].name]=df[i].isna().sum()
        index_numbers[i]=df[i].isna().sum()
dict_nulls
```

```python
df.dropna(axis=1, thresh=150,inplace=True)
df
```

```python
df.dropna(axis=0, thresh=80,inplace=True)
df
```

```python
for i in df.columns:
    if((isinstance(df[i].iat[0],float)==False) and(any(value.isalpha() for value in df[i].iat[0])==False) ):
        df[i]=df[i].astype(float)
for i in df.columns:
    if(df[i].dtypes!=object):
        df[i].fillna(df[i].median(), inplace=True)
df.info()
```

```python
df.to_csv('Cleaned_df.csv')
```
