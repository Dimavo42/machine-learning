```python
import requests
from bs4 import BeautifulSoup as bs
import re
import time
from collections import Counter
import pandas as pd
```

```python
page='https://worldpopulationreview.com/countries'
respones=requests.get(page)
page=bs(respones.text,'html.parser')
page_names=[] ## list of possible list name
all_possible_hrefs=[] ## list of possible links in the site
possible_dict={} ## Dict of all countries Subject:link
temp=[]## Tempory array to help remove the duplictes that resulted from scrapping helps order the dict if dict not needed may be removed
for link in page.find_all('section'):
    if((link.find('div',{'class':'actionbar py-1 px-2'})) and not(link.find('table')) and not(link.find('div',{'class':'letter-list'})) 
      and not link.find('ol') ):
        page_name=link.find('div',{'class':'actionbar py-1 px-2'})
        page_names.append(page_name.div.text)
        possible_dict[page_name.div.text]=[]
        for possible_urls in link.find_all('a'):
            all_possible_hrefs.append(possible_urls.get('href'))
            temp.append(possible_urls.get('href'))
    remove_dup=list(set(temp))
    if(temp):
        possible_dict[page_name.div.text].extend(remove_dup)
    
        
unique = list(set(all_possible_hrefs))
print(unique)



```

```python
#we Want to make a dict of all countryies so it would be easier to pull information
page='https://worldpopulationreview.com'
help_country_arr=[]
respones=requests.get(page)
page=bs(respones.text,'html.parser')
possible_dict={} ## Dict of all countries Subject:link
table=page.find('tbody',{'class':'jsx-2006211681'})
for row in table.find_all('tr'):
    if 'Rank' in row.a.text: 
        continue
    possible_dict[row.a.text]={}
    help_country_arr.append(row.a.text)
        
```

```python
page='https://worldpopulationreview.com/countries'
respones=requests.get(page)
page=bs(respones.text,'html.parser')
link_list={}
#We dont need all the information from the site only revlevent for the project so we select what information to take
for link in page.find_all('section'):
    if(link.find('div',{'class':'actionbar-title'})):
        if 'Country Articles' in link.find('div',{'class':'actionbar-title'}).text:
            for item in link.find_all('div',{'class':'article-list-item'}):
                if'Deaths per Day' in item.h3.text:
                    #link_list.append(item.a.get('href'))
                    link_list[item.h3.text]=item.a.get('href')
                if'Life Expectancy' in item.h3.text:
                    link_list[item.h3.text]=item.a.get('href')
                if'World Countries by National Debt' in item.h3.text:
                    link_list[item.h3.text]=item.a.get('href')
        if 'Country Listing Articles: Crime/Safety' in link.find('div',{'class':'actionbar-title'}).text:
            for item in link.find_all('div',{'class':'listing-grid-item'}):
                link_list[item.h5.text]=item.a.get('href')
        if 'Country Listing Articles: Economics' in link.find('div',{'class':'actionbar-title'}).text:
            for item in link.find_all('div',{'class':'listing-grid-item'}):
                link_list[item.h5.text]=item.a.get('href')
        if 'Country Listing Articles: Health' in link.find('div',{'class':'actionbar-title'}).text:
            for item in link.find_all('div',{'class':'listing-grid-item'}):
                link_list[item.h5.text]=item.a.get('href')


        
        
     
            
```

```python
#for one country ... 
page='https://worldpopulationreview.com'
respones=requests.get(page+link_list[0])
page=bs(respones.text,'html.parser')
time.sleep(2)
table=page.find('tbody',{'class':'jsx-2006211681'})
name_page=(page.find('div',{'class':'column is-9 is-clearfix py-6'})).h3.text
for row in table.find_all('tr'):
    for tab in table.find_all('td'):
        if(tab.text in help_country_arr):
            possible_dict[tab.text].update({name_page:tab.find_next("td").text})
            

print(possible_dict)
        
            
            
            

```

```python
for page_name,link in link_list.items():
    page='https://worldpopulationreview.com'
    respones=requests.get(page+link)
    page=bs(respones.text,'html.parser')
    time.sleep(5)
    table=page.find('tbody',{'class':'jsx-2006211681'})
    if(table):
        pass## there is a Table in the href so we take the first
    else:
        continue ## if there no table theres nothing to scrap in the href so we continue the loop 
    for row in table.find_all('tr'):
        for tab in table.find_all('td'):
            try:
                if(tab.text in help_country_arr):
                    possible_dict[tab.text].update({page_name:tab.find_next("td").text})
            except Exception as e:
                possible_dict[tab.text].update({page_name:None})
    print(page_name, link)
```
